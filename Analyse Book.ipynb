```python
import string
```

```python
def process_line(line, hist):
    line = line.replace('-',' ')
    
    for word in line.split():
        word = word.strip(string.punctuation + string.whitespace)
        word = word.lower() 
        hist[word] = hist.get(word, 0) + 1
```

```python
def process_file(filename):
    hist = {}
    fp = open(filename)
    
    for line in fp:
        process_line(line, hist)
    return hist
```

```python
def subtract(d1, d2):
    res = {}
    for key in d1:
        if key not in d2:
            res[key] = None
    return res
```

```python
def most_common(hist):
    t = []
    for key, value in hist.items():
        t.append((value, key))
    t.sort()
    t.reverse()
    return t
```

```python
hist = process_file('/root/emma.txt')
t = most_common(hist)
for freq, word in t[:25]:
    print(word, '\t', freq)
```

```python
words = process_file('/root/words.txt')
```

```python
diff = subtract(hist, words)
```
