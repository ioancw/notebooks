```python
%%time
import random
from math import exp, sqrt
```

```python
%%time
s_0 = 100
r = 0.05
T = 1.0
sigma = 0.2

values = []
for _ in range(1000000):
    s_T = s_0 * exp((r - 0.5 * sigma **2) * T +
        sigma * random.gauss(0,1) * sqrt(T))
    values.append(s_T) 
```
